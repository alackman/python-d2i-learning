# Conditionals

```python
df = pd.DataFrame({'ChildId':['id1', 'id2', 'id3', 'id4', 'id5'],
                   'Age first contact':[6,12,11,1,19],
                   'Gender':['M','m', 'F', '', 'F' ],
                   'Birthday':['01/01/2002', '02/02/2003', pd.NA, '03/03/2023', '06/01/2012'],
                   'Age':[5,12,11,6,2],
                   'CP Plan?':['Y', 'n', 'N', 'No', 'yES'],})
df['Birthday'] = pd.to_datetime(df['Birthday'], dayfirst=True)
df
```

```python
condition_over10 = df['Age'] >= 10

over_10 = df[condition_over10]

print(over_10)
```

```python
condition_under10_contact = df['Age first contact'] < 10

under_10_contact = df[condition_under10_contact]

print(under_10_contact)
```

```python
error_condition = df['Age first contact'] > df['Age']

error_df = df[error_condition]

error_df
```

```python
statement_1 = 1 > 2
statement_2 = 8/4 == 2

print((statement_1 | statement_2))
print((statement_1 & statement_2))
```

```python
# Divisible by 3 = fizz 
# Divisible by 5 = bang
# Both = fizzbang

numbers = list(range(1,101))

fizz = 3
bang = 5

for number in numbers:
    if (number % fizz == 0) & (number % bang == 0):
        print(f'{number} - Fizzbang')
    elif number % bang == 0:
        print(f'{number} - Bang')
    elif number % fizz == 0:
        print(f'{number} - Fizz')
    else:
        print(f'{number} - Poof')
```
